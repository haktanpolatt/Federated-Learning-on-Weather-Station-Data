# Kütüphanelerin İçe Aktarılması
Gerekli kütüphaneleri (pandas, numpy, matplotlib, seaborn) içe aktarın.

```python
# Kütüphanelerin İçe Aktarılması
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Veri Setinin Yüklenmesi
CSV dosyasını pandas kullanarak yükleyin ve ilk birkaç satırı görüntüleyin.

```python
# Veri Setinin Yüklenmesi
file_path = "dataset_synop.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1", delimiter=";")
df.head()
```

# Veri Setinin Genel İncelemesi
Veri setinin boyutunu, sütun isimlerini, veri türlerini ve eksik değerleri analiz edin.

```python
# Veri Setinin Genel İncelemesi
print("Veri setinin boyutu:", df.shape)
print("\nSütun isimleri:", df.columns)
print("\nVeri türleri:\n", df.dtypes)
print("\nEksik değerlerin sayısı:\n", df.isnull().sum())
```

# Eksik Verilerin Analizi ve Doldurulması
Eksik verileri analiz edin ve uygun yöntemlerle (ortalama, medyan, mod veya interpolasyon) doldurun.

```python
# Eksik Verilerin Analizi ve Doldurulması
# Eksik değerlerin yüzdesini hesaplayın
missing_percentage = df.isnull().sum() / len(df) * 100
print("Eksik değerlerin yüzdesi:\n", missing_percentage)

# Eksik değerleri doldurma (örnek: ortalama ile doldurma)
df["Sea level pressure"].fillna(df["Sea level pressure"].mean(), inplace=True)
df["Temperature"].fillna(df["Temperature"].mean(), inplace=True)
```

# Veri Türlerinin Dönüştürülmesi
Tarih veya kategorik sütunları uygun veri türlerine dönüştürün.

```python
# Veri Türlerinin Dönüştürülmesi
# Tarih sütununu datetime formatına dönüştürme
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%dT%H:%M:%S", errors="coerce")

# Kategorik sütunları dönüştürme
categorical_columns = ["Present weather", "Barometric trend type"]
for col in categorical_columns:
    df[col] = df[col].astype("category")
```

# Özellik Seçimi
Modelleme için gereksiz sütunları kaldırın ve önemli özellikleri seçin.

```python
# Özellik Seçimi
# Gereksiz sütunları kaldırma
columns_to_drop = ["Coordinates", "Name", "Barometric trend type.1"]
df.drop(columns=columns_to_drop, inplace=True, axis=1)

# Önemli özelliklerin seçimi
selected_features = ["Sea level pressure", "Temperature", "Humidity", "Date"]
df = df[selected_features]
```

# Veri Normalizasyonu ve Standardizasyonu
Numerik sütunları normalize edin veya standardize edin.

```python
# Veri Normalizasyonu ve Standardizasyonu
from sklearn.preprocessing import StandardScaler

# Standardizasyon
scaler = StandardScaler()
numerical_columns = ["Sea level pressure", "Temperature", "Humidity"]
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print("Standardize edilmiş veri:\n", df.head())
```
